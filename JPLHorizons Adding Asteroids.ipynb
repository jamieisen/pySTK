Install the JPL Horizons File Reader to STK

1. Download https://agiweb.secure.force.com/faqs/articles/HowTo/get-and-load-asteroid-ephemeris
2. Copy the XML file in the zip downloaded
3. Paste in AGI STK / Plugins folder
4. Right click on cmd prompt and run as administrator 
5. Change directory to the zip folder and run this line

regsvr32 JplHorizonsFileReader.wsc

Obtain ephemeris from JPL HORIZONS Web-Interface
https://ssd.jpl.nasa.gov/horizons.cgi

1. Ephemeris Type - VECTORS
2. Target Body - search an asteroid
3. Coordinate Origin - Geocentric [500]
4. Time Span - select a time span for scenario
5. Table Settings - see the screenshot
6. Display/Output - download/save


7. The resulting document will be named horizons_results, rename to asteroid name

```python
from pySTK import *
from AstroSat import *

# Initialize Scenario
stk = pySTK()
startEpoch = '11 Oct 2024 00:00:00.000'
BennuTest = stk.Scenario('BennuTest',startEpoch,'+1yr')
stk.InterplaneteryMode()
stk.SolarSystem()
stk.Asteroid('Bennu','C:\\Users\\jamie\\Desktop\\pySTK-Jupy\\bennu_results.txt',0.2625)
```
